```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def show_masks(img, masks_list):
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')

    for m in masks_list:
        # Support both SAM-style dict masks and plain numpy masks
        seg = m['segmentation'] if isinstance(m, dict) else m
        plt.imshow(np.ma.masked_where(~seg, seg), alpha=0.45)

    plt.show()  # show the final result
```

```python
from sam_predictor import predict_masks
```

```python
masks = predict_masks("data/images/00a52cd2a.jpg")
```

```python
img = cv2.imread("data/images/00a52cd2a.jpg")[:, :, ::-1]  # type: ignore
show_masks(img, masks)
```
