```python
# inference.ipynb
import os
import numpy as np
from tqdm import tqdm

# Person A utilities
from utils import get_rle_from_2d_mask, write_submission_csv
# Person B predictor
from sam_predictor import predict_masks
# Person C empty classifier
from is_empty_filter import is_empty

# Directory containing images (Person A renamed Kaggle train_v2 to data/images)
IMAGE_DIR = "data/images"

results = []

for img_name in tqdm(os.listdir(IMAGE_DIR)):
    img_path = os.path.join(IMAGE_DIR, img_name)

  
    if is_empty(img_path):
        results.append((img_name, []))
        continue


    masks = predict_masks(img_path)


    rles = [get_rle_from_2d_mask(m) for m in masks]

 
    results.append((img_name, rles))

write_submission_csv(results, "submission.csv")

```
